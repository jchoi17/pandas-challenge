```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(data)

# check to see file has been read
purchase_data
```

## Player Count

* Display the total number of players


```python
#using value_counts to get the number of unique players
total_players = len(purchase_data['SN'].unique())

#show total players
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# creating variables to hold values
items = len(purchase_data['Item Name'].unique())
num_purchase = purchase_data['Item Name'].count()
total_revenue = (purchase_data['Price'].sum()).astype(float)
avg_price = float(purchase_data['Price'].mean())

# putting variables into a summary table
summary_df = pd.DataFrame({'Number of Items': [items], 
                           'Number of Purchases':[num_purchase],
                           'Average Price': [avg_price],
                          'Total Revenue': [total_revenue], 
                          'Number of Players': [total_players]})

# adding in dollar signs
summary_df['Total Revenue'] = summary_df['Total Revenue'].map("${:.2f}".format)
summary_df['Average Price'] = summary_df['Average Price'].map("${:.2f}".format)

#show table
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# calculate percentage of genders
gender_df = purchase_data.groupby('Gender')
members_by_gender = gender_df['SN'].nunique()
percent_gender = ((members_by_gender/total_players)*100).map("{:,.2f}%".format)

# create gender summary table
gender_summary = pd.DataFrame({'Player Count': members_by_gender,
                               'Percentage by Gender': percent_gender})

#show table
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# create new variables to hold values by gender
total_purchase_count = gender_df['Purchase ID'].count()
total_purchase_value = gender_df['Price'].sum()
avg_purchase_gender = gender_df['Price'].mean()
avg_total_person = total_purchase_value/members_by_gender

                                                                
# create dataframe for summary table
purchases_by_gender = pd.DataFrame({'Purchase Count': total_purchase_count,
                                   'Average Purchase Price': avg_purchase_gender,
                                   'Avg Total Purchase per Person': avg_total_person,
                                   'Total Purchase Value': total_purchase_value})
# map values to dollar signs
purchases_by_gender['Average Purchase Price'] = purchases_by_gender['Average Purchase Price'].map("${:.2f}".format)
purchases_by_gender['Avg Total Purchase per Person'] = purchases_by_gender['Avg Total Purchase per Person'].map("${:.2f}".format)
purchases_by_gender['Total Purchase Value'] = purchases_by_gender['Total Purchase Value'].map("${:.2f}".format)

# show table
purchases_by_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create bins & name them
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# create column to hold bins
purchase_data['Age Summary'] = pd.cut(purchase_data['Age'], bins, labels=group_names, include_lowest=True)

# groupby
age_groups = purchase_data.groupby('Age Summary')

#drop duplicates
members_by_age = age_groups['SN'].nunique()

# run equations
percent_by_age = ((members_by_age/total_players)*100).map("{:,.2f}%".format)

#create dataframe for ages
age_summary = pd.DataFrame({'Total Count': members_by_age, 
                         'Percentage of Players': percent_by_age})

# show table
age_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by SN 
members = purchase_data.groupby('SN')

#run counts/formulas
count_member = members['Purchase ID'].count()
value_member = members['Price'].sum()
avg_member = members['Price'].mean()

#create dataframe
top_spenders = pd.DataFrame({"Purchase Count": count_member,
                             "Average Purchase Price": avg_member,
                             "Total Purchase Value":value_member})
#sort to get top spenders
top_spenders = top_spenders.sort_values(["Total Purchase Value"], ascending=False).head()

#add mapping to format it
top_spenders['Average Purchase Price'] = top_spenders['Average Purchase Price'].map("${:.2f}".format)
top_spenders['Total Purchase Value'] = top_spenders['Total Purchase Value'].map("${:.2f}".format)

#show table
top_spenders
```
